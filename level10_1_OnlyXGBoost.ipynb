# XGBoost with OneHotEncoder

```python
from sklearn.preprocessing import OneHotEncoder

data2.reset_index(drop=True, inplace=True)
encoder = OneHotEncoder(sparse=False)
encoded_data = encoder.fit_transform(data2[['POWER', 'UNIT', 'MODE']])
# Create a new DataFrame with the encoded data
df_encoded = pd.DataFrame(encoded_data, columns=encoder.get_feature_names(['POWER', 'UNIT', 'MODE']))

# Concatenate the original DataFrame and the encoded DataFrame
result = pd.concat([data2, df_encoded], axis=1)
# df = df.drop(['UNIT'], axis=1)
data2 = result.drop(['MODE', 'POWER', 'UNIT'], axis=1)
data2
```

```python
df = data2
X = df
Y = X['OUTPUT'].values
X = X.drop('OUTPUT', axis = 1).values
```

```python
xgb_model = xgb.XGBRegressor(objective="reg:squarederror", random_state=0)
```

```python
xgb_model.fit(X, Y)

y_pred = xgb_model.predict(X)
mse=mean_squared_error(Y, y_pred)

print(np.sqrt(mse))
```

```python
file_path = "./data/test_data.csv"
df2 = pd.read_csv(file_path)
df2['POWER_NUM'] = df2["VOLTS"]*data["AMPS"]

# Transform the test data using the same encoder instance
encoded_test_data = encoder.transform(df2[['POWER', 'UNIT', 'MODE']])

# Create a new DataFrame with the encoded test data
df_encoded_test  = pd.DataFrame(encoded_test_data, columns=encoder.get_feature_names(['POWER', 'UNIT', 'MODE']))
df2 = pd.concat([df2, df_encoded_test], axis=1)

df2 = df2.drop(['MODE', 'POWER', 'UNIT'], axis=1)
df2
```

```python
y_pred = xgb_model.predict(df2)
np.savetxt('./output/output_XGBoostOneHotEncoding.csv', y_pred)
```

# XGBoost with categorical

```python
df = base_df.copy()
df["UNIT"] = df["UNIT"].astype("category")
df["MODE"] = df["MODE"].astype("category")
df["POWER"] = df["POWER"].astype("category")
X = df
y = X['OUTPUT']
X = X.drop('OUTPUT', axis = 1)
```

```python
xgb_model = xgb.XGBRegressor(tree_method="hist", objective="reg:linear", random_state=0, enable_categorical=True)
```

```python
X
```

```python
xgb_model.fit(X, y)

y_pred = xgb_model.predict(X)
mse=mean_squared_error(y, y_pred)

print(np.sqrt(mse))
```
